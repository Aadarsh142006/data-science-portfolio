```python
### try,except,else block
try:
    num=int(input("enter a num"))
    result=1/num
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
## try,except,else and finally
try:
    num=int(input("enter a num"))
    result=1/num
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    print("exceution completed")
```
