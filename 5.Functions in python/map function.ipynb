```python
numbers=[1,2,3,4,5,6]
list(map(lambda x:x**2,numbers))
```

```python
### map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]
added_num=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_num)
```

```python
### map() to convert a list of strings to integer
str_num=['1','2','3']
int_num=list(map(int,str_num))
print(int_num)
```

```python
words=["apple","banana","cherry"]
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person["name"]
people=[{"name":"shiva","age":20},
{"name":"jack","age":30}
]
list(map(get_name,people))
```
