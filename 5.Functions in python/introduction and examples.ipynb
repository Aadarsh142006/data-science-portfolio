```python
### variable length arguments
### positional and keywords argumenst
 
def print_numbers(*args):# *aadi is used for number of arguments
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"aadi")##positional arguments 
```

```python
### keyword arguments

def print_details(*args,**aada):
    for number in args:
        print(f" positional arguments :{number}")
        
    for key,value in aada.items():
        print(f"{key}:{value}")
```

```python
print_details(name="krish",age=32,country="india")
```

```python
print_details(1,2,3,4,5,6,"aadarsh",name="krish",age=32,country="india")
```

```python
def convert_temperature(temp, unit):
    if unit == "C":  # Celsius to Fahrenheit
        return (temp * 9/5) + 32
    elif unit == "F":  # Fahrenheit to Celsius
        return (temp - 32) * 5/9
    else:
        return "Invalid unit. Use 'C' or 'F'."

# Example usage:
print(convert_temperature(100, "C"))  # 100°C → 212°F
print(convert_temperature(212, "F")) # 212°F → 100°C
```

```python
## password strength checker
def password_checker(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^*" for char in password):
        return False
    else:
         return True
print(password_checker("aadarsH@123"))
    
    
```

```python
## calculate the total cost of items in a shopping mall
def cal_tot_cost(cart):
    totalcost=0
    for item in cart:
        totalcost+=item["price"]*item["quantity"]
    return totalcost

cart=[
    {"name":"apple","price":200,"quantity":7},
    {"name":"bag","price":500,"quantity":8},
    {"name":"pen","price":20,"quantity":9},
]
print(cal_tot_cost(cart))
```

```python
def fac(num):
    if num==0 or num ==1:
        return 1
    else:
        return num*fac(num-1)
print(fac(4))
```

```python
## read a file and count the frequency of each word

def count_word_freq(file_path):
    wordcount={}
    with open(file_path,"r") as file:
         for line in file:
             words=line.split()
             for word in words:
                 word=word.lower().strip(".,!?/\\")
                 wordcount[word]=wordcount.get(word,0)+1
    return wordcount
file_path="sample.txt"
word_freq=count_word_freq(file_path)
print(word_freq)

```

```python
def valid_email_address(email):
    # Must contain '@'
    if "@" not in email:
        return False
    # Must contain '.'
    if "." not in email:
        return False
    # No spaces allowed
    if " " in email:
        return False
    # Must end with '.com'
    if not email.endswith(".com"):
        return False
    return True

# Example usage
print(valid_email_address("aadarshnanna08@gmail.com"))  # True
print(valid_email_address("user@domain"))              # False
print(valid_email_address("user name@domain.com"))     # False
```
