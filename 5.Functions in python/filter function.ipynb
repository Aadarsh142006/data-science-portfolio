## filter() is function constructor an iterator from elements of an iterable for which returns true.it is used to fliter out items from a list or any iterable based ona condition

```python
lst=[1,2,3,4]
list(filter(lambda x:x%2==0,lst))
list(filter(lambda x:x>2,lst))
```

```python
## filter with mul condition
list(filter(lambda x:x>1 and x%2==0,lst))
```

```python
people = [
    {"name": "shiva", "age": 20},
    {"name": "jack", "age": 30}
]

# Filter people whose age > 25
result = list(filter(lambda person: person["age"] > 25, people))
print(result)
```
