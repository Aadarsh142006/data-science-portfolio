```python
## a function with out name is lambda function
## it an have any no of arguments but only expression
## syntax
lambda arguments:expression
```

```python
def addi(a,b):
    return a+b
addi(3,4)
```

```python
addition=lambda a,b:a+b
type(addition)
print(addition(4,5))
```

```python
## even or odd
even=lambda num:num%2==0
print(even(4))
```

```python
add=lambda a,b,c:a+b+c
print(add(1,2,3))
```
