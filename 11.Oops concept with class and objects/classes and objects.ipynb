```python
### modelling a bank account

## define a class for bank acc
class Bank:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        if amount>0:
            self.balance+=amount
            print(f"{amount} is deposited in to {self.owner} and balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("amount insufficent")
        else:
            self.balance-=amount
            print(f"your balance is now {self.balance}")
    def getbalance(self):
        return self.balance
### creating objects 
account=Bank("tabrez",30000)
print(account.balance)

    
    
```
