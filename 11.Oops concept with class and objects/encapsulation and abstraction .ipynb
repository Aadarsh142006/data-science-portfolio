Encapsulation is the concept of wrapping the data and methods together as a single unit.It restricts direct access to some of the objects components,which is a means of preventing accidental interface and misuse of the data

```python

## public,private,protected variabiles

class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age  ##public variables
person=Person("a",30)
person.name
dir(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name
        self.__age=age  ##pravite variables
        self.gender=gender

def get_name(person):
    return Person.__name

person=Person("a",30,'m')
get_name(person)

dir(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self._name=name
        self._age=age  ##protected variables can be accesed with only derived classes only not all
        self.gender=gender
class Emp(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

emp=Emp("a",30,'m')
emp._name

```

```python
### encapsulation with getter and setter methods
class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age  ##pravite variables

    ## getter method for name
    def get_name(self):
        return self.__name

    ## setter method for name
    def set_name(self,name):
        self.__name=name

     ## getter method for name
    def get_age(self):
        return self.__age

    ## setter method for name
    def set_age(self,age):
        if age >0:
            self.__age=age



person=Person("aadarsh",30,)

## access and modify private variables using getter and setter
print(person.get_name())
print(person.get_age())

print(person.set_age(20))
print(person.get_age())
```

```python
### abstraction hidding the data compplex showing necessary things 
from abc import ABC,abstractmethod

## abstract base class
class Vehicle(ABC):
    def drive(self):
        print("driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("car started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)
```
