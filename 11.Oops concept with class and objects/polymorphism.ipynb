```python
### methods overriding
class Animal:
    def speak(self):
        return "sound of animal"
class Dog(Animal):
    def speak(self):
        return 'wooff woof!'
class Cat(Animal):
    def speak(self):
        return 'meow mewo!!'
def animal_speak(animal):
    print(animal.speak())
    
buddy=Dog()
buddy.speak()
cat=Cat()
cat.speak()
animal_speak(buddy)
animal_speak(cat)
        
```

```python
#### plomorphism with functions and methods
## base class
import math
class Shape:
    def area(self):
        return "the area of the figure"
## derived class
class Rectangle(Shape):
    def __init__(self,width,length):
        self.width=width
        self.length=length
    def area(self):
        return self.width*self.length

## derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return math.pi*self.radius*self.radius
def print_area(Shape):
    print(f'the area is {Shape.area()}')
        
recc=Rectangle(4,5)
print_area(recc)
cir=Circle(10)
print_area(cir)
```

```python
### interface in python called Abstract base class
from abc import ABC,abstractmethod

### define abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    ## deriveed class 1
class Car(Vehicle):
    def start_engine(self):
        return "car engine started"
    
## derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "motorcycle engine started"
    
## create objects of car and motorcyle
car=Car()
bike=Motorcycle()
car.start_engine()
bike.start_engine()

```
