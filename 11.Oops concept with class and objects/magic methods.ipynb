```python
## basic methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"person(name={self.name},age={self.age})"

person=Person("aadarsh",20)
print(person)
print(repr(person))
```

```python
## overloading done by operating functions
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    def __mul__(self,other):
        return Vector(self.x*other.x,self.y*other.y)
    def __eq__(self,other):
        return Vector(self.x==other.x,self.y==other.y)
    def __repr__(self):
        return f"Vector({self.x},{self.y})"
    
    v1=Vector(2,4)
    v2=Vector(3,5)
    v3=v1+v2
    print(v3)
    print(v1-v2)
    print(v1*v2)
    print(v1==v2)
```
