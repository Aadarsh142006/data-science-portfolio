custom exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass
class dobException(Error):
    pass
```

```python
year=int(input("enter your dob of year"))
age=2026-year
try:
    if age<=30 and age>=20:
        print("the age is valid you can apply for the exam")
    else:
        raise dobException
except dobException:
    print("sorry your age greater than 20 and less than 30")
```
