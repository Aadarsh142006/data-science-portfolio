```python
def generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

```python
### practical : reading large files
def read_large_file(file_path):
    with open(file_path,'r')as file:
        for line in file:
            yield line


```

```python
file_path='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())

```
