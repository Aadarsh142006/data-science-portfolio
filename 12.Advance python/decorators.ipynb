A decorator in Python is a special function that lets you add extra behavior to another function without changing its code.
Think of it like wrapping a gift: the gift (your function) stays the same, but the wrapping (the decorator) adds something extra.
- A decorator takes a function as input.
- It returns a new function with modified or extended behavior.
- You apply it using the @decorator_name syntax above your function.

```python
### closures
def main_method(func,lst):
    a="banni"
    print("aadarsh")
    def second_method():
        print("sai")
        print(func(lst))
    return second_method()

```

```python
main_method(len,[1,2,3,4,5])
```

```python
def my_decorator(func):
    def wrapper():
        print("a")
        func()
        print('b')
    return wrapper()
```

```python
@my_decorator
def say_hello():
    print("hello !")
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
