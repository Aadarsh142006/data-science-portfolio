Example 2: Organizing Student Grades

```python
## organizing the grades
grades=[85,90,78,90,89]

## adding the new grade
grades.append(98)

## calculating the avergae grade
avge=sum(grades)/len(grades)
print(avge)

## finding the highests and lowest grades
print(f"{ "minimum"}:{min(grades)}")
print(f"{ "maximum"}:{max(grades)}")
```
