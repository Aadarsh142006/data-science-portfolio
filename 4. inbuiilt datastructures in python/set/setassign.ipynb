```python
set1=set(x for x in range(11))
print(set1)
```

```python
set1.add(11)
set1.remove(1)
print(set1)
```

```python
s1=set(x for x in range(5))
s2=set(x for x in range(10) if x%2==0)
print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2))
```

```python
s3=set(x**2 for x in range(11))
print(s3)
```

```python
print(set(x for x in s3 if x%2==0))
```

```python
s4=set()
print(s1.issubset(s4))
print(s1.superset(s4))
```
