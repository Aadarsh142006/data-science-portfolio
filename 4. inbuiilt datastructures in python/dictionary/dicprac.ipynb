```python
##accesing dictionary elements
student={"name":"aadarsh","age":20,"grade":"A"}
print(student["grade"])
print(student["name"])
#accesing using get() method
print(student.get("age"))
print(student.get("aadarsh"))
print(student.get("aadarsh","Not avaiabile"))

```

```python
##modifying dict elelments
#dict is mutable but key is unique
student["age"]=100#updating values 
print(student)
student["corse"]="ml" #add new key and value
print(student)
del student["grade"]#delete key and value pair
print(student)
```

```python
#dict methods 
keys=student.keys()#gives the all keys
print(keys)

values=student.values()#all values
print(values)
items=student.items()#it returns key value pairs in list or tuple 
print(items)
```

```python
#shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="shiva"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()##shallow copy
print(student_copy1)
print(student)
student["name"]="aadarsh"
print(student)
print(student_copy1)
```

```python
##iterating over dicitionary 
##try can use loops to iterate over dict or items

#iterating over keys 
for key in student.keys():
    print(key)
    ##iterate over values
    for value in student.values():
        print(value)
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
##nested dictionary
students={
    "student1":{"name":"shiva","age":19},
    "student2":{"name":"peter","age":20}

}
print(students)
```

```python
##acces the nested dictionary
print(students["student1"]["name"])
```

```python
#iterating over nested dictionaries
for  student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
##dict comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dict comprehension
even={x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
## practical examples
numbers=[1,2,3,4,3,4,2,3,5,6,6]
fre={}
for num in numbers:
    if num in fre:
        fre[num]+=1
    else:
        fre[num]=1
print(fre)
```

```python
# merge two dictionaries
dict1={"a":1,"b":2}
dict2={"b":4,"c":5}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
