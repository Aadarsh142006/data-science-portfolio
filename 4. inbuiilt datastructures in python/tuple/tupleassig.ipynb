```python
# Step 1: Create a tuple with duplicate elements
tup = (1, 2, 3, 2, 4, 1, 5)

# Step 2: Convert tuple to set (removes duplicates automatically)
unique_set = set(tup)

# Step 3: Print the resulting set
print(unique_set)
```

```python
# Functions to compute min, max, and sum of a tuple
def get_min(tup):
    return min(tup)

def get_max(tup):
    return max(tup)

def get_sum(tup):
    return sum(tup)

# Sample tuple
sample_tup = (10, 4, 7, 2, 15)

# Print results
print("Tuple:", sample_tup)
print("Minimum:", get_min(sample_tup))
print("Maximum:", get_max(sample_tup))
print("Sum:", get_sum(sample_tup))
```
