```python
#packing and unpacking
packed_tuple=1,"aadarsh",3.14#by deafult it is packed as tuple when you mentioned as commas
print(packed_tuple)
```

```python
a,b,c=packed_tuple#it is unpacking
print(a)
print(b)
print(c)
```

```python
#unpacking with *
num=(1,2,3,4,5,6,7)
first,*middle,last=num
print(first)
print(middle)
print(last)
```

```python
##nested tuple 
##nested list
lst1=[[1,2,3],[39],("hello")]
print(lst1[2])
```

```python
tup1=((1,2,3),(39,),("hello"))#(39) is not utple where (39,)is tuple 
print(tup1[0])
print(tup1[2])
```

```python
#iterating the tuple
for sub in tup1:
    for item in sub:
        print(item,end="")
    print()
```
