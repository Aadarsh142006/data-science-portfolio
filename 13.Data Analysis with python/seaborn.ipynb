```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
## Basic plotting using seaborn
tips=sns.load_dataset('tips')
tips
```

```python
### create a scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('first plot')
plt.show()
```

```python
## line plot
sns.lineplot(x='sex',y='tip',data=tips)
plt.title('first line plot')
plt.xlabel('sex')
plt.ylabel('tip')
plt.grid(True)
plt.show()
```

```python
## categorical plots
## bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('first bar plot')
plt.show()
```

```python
## box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('first box plot')
plt.show()
```

```python
### violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('first violen plot')
plt.show()
```

```python
## histogram plot

sns.histplot(tips['total_bill'],bins=10,kde=True,color='green')
plt.title('this is my first histogram plot')
plt.show()
```

```python
### kde plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## pair plot 
sns.pairplot(tips)
```

```python
### heat map
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
