```python
import sqlite3
connection=sqlite3.connect('example.db')
```

```python
cursor=connection.cursor()
```

```python
## create table

cursor.execute('''
create table if not exists employee(
   id integer primary key,
   name text not null,
   age integer,
   department text
)
''')

### commit the changes 
connection.commit()
```

```python
cursor.execute("SELECT * FROM employee")
print(cursor.fetchall())
```

```python
### inserting 
cursor.execute('''
insert into employee(id,name,age,department)
values(121,'aadarsh',20,'aids')

''')

cursor.execute('''
insert into employee(id,name,age,department)
values(120,'mohith',20,'aids')

''')

cursor.execute('''
insert into employee(id,name,age,department)
values(113,'chiru',20,'aids')

''')
connection.commit()
```

```python
## update the data in the table
cursor.execute('''
update employee
set age=24
where name='mohith'
''')
connection.commit()
```

```python
cursor.execute('SELECT * FROM employee')
print(cursor.fetchall())
```

```python
cursor.execute('''
delete from employee
where name='aadarsh'
''')
connection.commit()
```

```python
### working with the strucured data

connection=sqlite3.connect('sql_data..db')
cursor=connection.cursor()
```

```python
cursor.execute('''
create table if not exists sales(
id integer primary key,
   name text not null,
   age integer,
   department text
)

''')
connection.commit()
```

```python
sales_data={
    (121,'a',20,'aids'),
       (122,'f',20,'aids'),
          (123,'e',20,'aids'),
             (124,'d',20,'aids')
}
```

```python
cursor.executemany('''
insert into sales(id,name,age,department)
values(?,?,?,?)
''',sales_data)
connection.commit()
```

```python
cursor.execute('select * from sales')
print(cursor.fetchall())
```
