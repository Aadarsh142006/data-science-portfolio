```python
import logging

 ## configure the basic the logging setting
logging.basicConfig(level=logging.DEBUG)

## logging message
logging.debug("this is debug msg")
logging.info("this is info msg")
logging.error('this is error')
logging.warning("this is warning")
logging.critical("this is a critical msg")
```

```python
### configure logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
level=logging.DEBUG,
format='%(asctime)s-%(name)s-%(message)s',
datefmt="%y-%m-%d %H:%M:%S",
force=True
)
## logging message
logging.debug("this is debug msg")
logging.info("this is info msg")
logging.error('this is mistake')
logging.warning("this is warning")
logging.critical("this is a critical msg")
```
