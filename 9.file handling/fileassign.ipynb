```python
def write(file_path):
    with open(file_path,'w')as file:
        file.write("sai\n")
        file.write("mottu\n")
file_path='sample.txt'
write(file_path)
```

```python
def copy(source_path,destin_path):
    with open(source_path,'r')as file:
        content=file.read()
    with open(destin_path,'w')as file:
        file.write(content)
source_path='example.txt'
destin_path='sample.txt'
copy(source_path,destin_path)
```

```python
def count_word(file_path):
    with open(file_path,'r')as file:
        line=file.read()
        count=len(line.split())
    return count
     
print(count_word('sample.txt'))
```

```python
def replace(file_path,word,word2):
    with open(file_path,'r')as file:
        content=file.read()
    content = content.replace(word, word2)
    with open(file_path, 'w') as file:
        file.write(content)


file_path,word,word2='sample.txt','hello',"mottu"
replace(file_path,word,word2)
with open(file_path,'r')as file:
    content=file.read()
    print(content)


```

```python
def reverse(file_path):
    with open('sample.txt','r')as file:
        lines=file.readlines()
        for line in reversed(lines):
            print(line.strip())


```

```python
def merge(file_path,file1,file2):
    with open(file1,'r')as file:
        content1=file.read()
    with open(file2,'r')as file:
        content2=file.read()
    with open(file_path,'w')as file:
        file.write(content1)
        file.write(content2)
file_path,file1,file2='merge.txt','sample.txt','example.txt'
merge(file_path,file1,file2)
with open(file_path,'r')as file:
    print(file.read())

```
