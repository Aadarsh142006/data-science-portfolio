```python
### read a whole file
with open('example.txt',mode='r') as file:
    content=file.read()
    print(content)
```

```python
### read a file line by line
with open('example.txt','r')as file:
    for line in file:
        print(line.strip())#strip() removes new line character
```

```python
### writing a file
with open('example.txt','w') as file:
    file.write('hello world\n')
    file.write("this is new line\n")
```

```python
## wite a file without overriding
with open('example.txt','a')as file:
    file.write("appended \n")
```

```python
### writing list of lines to a file
lines=['first line\n.','second line\n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
### binary files
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb')as file:
    file.write(data)
```

```python
with open('example.bin',mode='rb') as file:
    content=file.read()
    print(content)
```

```python
### read the content from a source text file and write to a destination text file
# copying a text file
with open('example.txt','r')as source_file:
    content=source_file.read()
with open('destination.txt','w')as destiny_file:
    destiny_file.write(content)
```

```python
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        lines_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return lines_count, word_count, char_count

file_path = "example.txt"
lines, words, characters = count_text_file(file_path)

print(f"lines: {lines}, words: {words}, characters: {characters}")
```
